## Cluster 1

```python
#importer les packages kmeans
from sklearn.cluster  import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
dfa=df[['Income',"MntWines"]]
dfa
```

```python
#standardiser les données
scdfa=StandardScaler()
dfa_std=scdfa.fit_transform(dfa.astype(float))
dfa_std
```

```python
#Appliquer kmeans
cluster_range=[2,3,4,5,6,7,8,9,10]
inertie=[]

for c in cluster_range:
    kmeans=KMeans(init='k-means++',n_clusters=c).fit(dfa_std)
    inertie.append(kmeans.inertia_)
    
#print(inertie)

#dessiner les inerties calculé
plt.Figure()
plt.plot(cluster_range,inertie,marker='o')
plt.show()
```

```python
#On part sur 3 clusters
kmeans=KMeans(n_clusters=4).fit(dfa_std)

#affecter les cluster à la varialb e labels
labels=kmeans.labels_

#créer un dataset qui contient les valeurs de dfa_std
new_dfa=pd.DataFrame(data=dfa_std,columns=['Income','Mnt_Wines'])

#rajouter les labels à mon dataset dfa_std
new_dfa['labels_kmeans']=labels

new_dfa
```

```python
#visualiser les cluster en 2dimesions
fig,ax=plt.subplots(figsize=[10,7])
plt.scatter(new_dfa['Income'][new_dfa['labels_kmeans']==0],new_dfa['Mnt_Wines'][new_dfa['labels_kmeans']==0],
            color='blue',s=100,linestyle='--')
plt.scatter(new_dfa['Income'][new_dfa['labels_kmeans']==1],new_dfa['Mnt_Wines'][new_dfa['labels_kmeans']==1],
            color='red',s=100,linestyle='--')
plt.scatter(new_dfa['Income'][new_dfa['labels_kmeans']==2],new_dfa['Mnt_Wines'][new_dfa['labels_kmeans']==2],
            color='green',s=100,linestyle='--')

plt.xlabel('Depense Vin')
plt.ylabel('Income')
plt.show()
```

## Cluster 2

```python
#importer les packages kmeans
from sklearn.cluster  import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Création de l'encodeur
label_encoder = LabelEncoder()

# Encodage de la colonne
df['Education_Encoded'] = label_encoder.fit_transform(df['Education'])
```

```python
dfa2 =df[['Income', 'Education_Encoded']]
dfa2
```

```python
#standardiser les données
scdfa=StandardScaler()
dfa2_std=scdfa.fit_transform(dfa.astype(float))
dfa2_std
```

```python
#Appliquer kmeans
cluster_range=[2,3,4,5,6,7,8,9,10]
inertie=[]

for c in cluster_range:
    kmeans=KMeans(init='k-means++',n_clusters=c).fit(dfa2_std)
    inertie.append(kmeans.inertia_)
    
#print(inertie)

#dessiner les inerties calculé
plt.Figure()
plt.plot(cluster_range,inertie,marker='o')
plt.show()
```

```python
#On part sur 3 clusters
kmeans=KMeans(n_clusters=5).fit(dfa2_std)

#affecter les cluster à la varialb e labels
labels=kmeans.labels_

#créer un dataset qui contient les valeurs de dfa_std
new_dfa2=pd.DataFrame(data=dfa2_std,columns=['Income','Education'])

#rajouter les labels à mon dataset dfa_std
new_dfa2['labels_kmeans']=labels

new_dfa2
```

```python
#visualiser les cluster en 2dimesions
fig,ax=plt.subplots(figsize=[10,7])
plt.scatter(new_dfa2['Education'][new_dfa2['labels_kmeans']==0],new_dfa2['Income'][new_dfa2['labels_kmeans']==0],
            color='blue',s=100,linestyle='--')
plt.scatter(new_dfa2['Education'][new_dfa2['labels_kmeans']==1],new_dfa2['Income'][new_dfa2['labels_kmeans']==1],
            color='red',s=100,linestyle='--')
plt.scatter(new_dfa2['Education'][new_dfa2['labels_kmeans']==2],new_dfa2['Income'][new_dfa2['labels_kmeans']==2],
            color='green',s=100,linestyle='--')

plt.show()
```
