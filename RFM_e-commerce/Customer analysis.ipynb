```python
# Convertir les colonnes de dates en format datetime
DBC1['order_purchase_timestamp'] = pd.to_datetime(DBC1['order_purchase_timestamp'])
DBC1['order_delivered_customer_date'] = pd.to_datetime(DBC1['order_delivered_customer_date'])

DBC1['payment_installments'] = DBC1['payment_installments'].astype(str)

# Calculer le délai de livraison
DBC1['delais_de_livraison'] = DBC1['order_delivered_customer_date'] - DBC1['order_purchase_timestamp']

# Agréger par customer_unique_id
aggregated_df = DBC1.groupby('customer_unique_id').agg({
    'delais_de_livraison': 'first',  
    'payment_value': 'sum',
    'product_category_name_english': lambda x: ', '.join(x.unique().astype(str)),  
    'review_score': 'first',  
    'payment_installments': lambda x: ', '.join(x.unique().astype(str)),  
    'payment_type': lambda x: ', '.join(x.unique().astype(str)),  
    'customer_zip_code_prefix': 'first',
    'customer_city': 'first',
    'customer_state': 'first',
    'order_purchase_timestamp': 'first'  
}).reset_index()

# Calculer la récence
aggregated_df['Recency'] = pd.to_datetime('now') - aggregated_df['order_purchase_timestamp']

# Calculer la fréquence
frequency_df = DBC1.groupby('customer_unique_id')['order_id'].nunique().reset_index()
frequency_df.columns = ['customer_unique_id', 'Frequency']

# Fusionner la fréquence avec le DataFrame agrégé
aggregated_df = pd.merge(aggregated_df, frequency_df, on='customer_unique_id', how='left')

# Renommer la colonne 'payment_value' en 'Monetary'
aggregated_df.rename(columns={'payment_value': 'Monetary'}, inplace=True)

# Afficher le DataFrame agrégé
aggregated_df
```

```python
# Segmenter les clients en utilisant des quintiles
aggregated_df['R_Score'] = pd.qcut(aggregated_df['Recency'], 5, labels=range(5, 0, -1))
aggregated_df['F_Score'] = pd.qcut(aggregated_df['Frequency'], 5, labels=range(1, 6))
aggregated_df['M_Score'] = pd.qcut(aggregated_df['Monetary'], 5, labels=range(1, 6))

# Calculer le score RFM final
aggregated_df['RFM_Score'] = aggregated_df['R_Score'].astype(str) + aggregated_df['F_Score'].astype(str) + aggregated_df['M_Score'].astype(str)

# Afficher les résultats
print(aggregated_df.head())
```
